```python
import matplotlib.pyplot as plt

# Count the occurrences of each category
category_counts = df['Appliance Category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(category_counts, 
        labels=category_counts.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=plt.cm.Paired.colors)

plt.title('Distribution of Appliance Categories', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Create a blank figure
fig, ax = plt.subplots(figsize=(12, 3))

# Hide axes
ax.axis('tight')
ax.axis('off')

# Grab the first 5 rows for the table
table_data = df.head()

# Draw the table
table = ax.table(cellText=table_data.values, 
                 colLabels=table_data.columns, 
                 loc='center', 
                 cellLoc='center')

# Adjust layout spacing
table.scale(1, 1.8)
plt.title('Top 5 Rows of Appliance Data', fontsize=14, fontweight='bold', pad=20)
plt.show()
```
